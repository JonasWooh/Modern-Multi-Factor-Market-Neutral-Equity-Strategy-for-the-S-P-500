```python
from pathlib import Path
import gzip
import shutil

source_dir = Path(".")          # 当前目录
target_dir = source_dir / "csv" # 输出目录：./csv
target_dir.mkdir(exist_ok=True)

csv_gz_files = list(source_dir.glob("*.csv.gz"))

if not csv_gz_files:
    print("当前目录下未找到 .csv.gz 文件。")
else:
    for gz_file in csv_gz_files:
        # 例如：a.csv.gz -> a.csv
        out_file = target_dir / gz_file.with_suffix("").name
        with gzip.open(gz_file, "rb") as f_in, open(out_file, "wb") as f_out:
            shutil.copyfileobj(f_in, f_out)
        print(f"已转换: {gz_file.name} -> {out_file}")
```
